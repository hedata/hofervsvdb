```python
import tweepy #https://github.com/tweepy/tweepy
import csv
import json
onetweet ={}
#Twitter API credentials
consumer_key = "secret"
consumer_secret = "secret"
access_key = "secret"
access_secret = "secret"


def get_all_tweets(screen_name):
    #Twitter only allows access to a users most recent 3240 tweets with this method
    
    #authorize twitter, initialize tweepy
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_key, access_secret)
    api = tweepy.API(auth)
    
    #initialize a list to hold all the tweepy Tweets
    alltweets = []

    #make initial request for most recent tweets (200 is the maximum allowed count)
    new_tweets = api.user_timeline(screen_name = screen_name,count=200)
    
    #save most recent tweets
    alltweets.extend(new_tweets)
    
    #save the id of the oldest tweet less one
    oldest = alltweets[-1].id - 1
    
    #keep grabbing tweets until there are no tweets left to grab
    while len(new_tweets) > 0:
        print ("getting tweets before "+str(oldest))
        
        #all subsiquent requests use the max_id param to prevent duplicates
        new_tweets = api.user_timeline(screen_name = screen_name,count=200,max_id=oldest)
        
        #save most recent tweets
        alltweets.extend(new_tweets)
        #update the id of the oldest tweet less one
        oldest = alltweets[-1].id - 1
        print ("..."+str(len(alltweets))+" tweets downloaded so far")

    #transform the tweepy tweets into a 2D array that will populate the csv	
    #.encode("utf-8")re
    outtweets = []
    for tweet in alltweets:
        test = False
        try:
            test = tweet.retweeted_status
            test = True;
            print("found a retweet")
        except:
            test = False;
        #hashtags
        hashtags = ""
        #print(json.dumps(tweet.entities, sort_keys=True,indent=4, separators=(',', ': ')))
        for hashtag in tweet.entities['hashtags']:
            hashtags = hashtags+" "+hashtag['text']
        user_mentions = 0
        for use in tweet.entities['user_mentions']:
            user_mentions = user_mentions +1
        outtweets.append([tweet.id_str, tweet.created_at, tweet.text, tweet.entities
                 , tweet.favorite_count, tweet.retweet_count, 
                          tweet.user.followers_count,tweet.user.friends_count
                          ,tweet.retweeted,test,hashtags,user_mentions])
    #outtweets = [[tweet.id_str, tweet.created_at, tweet.text, tweet.entities
    #             , tweet.favorite_count, tweet.retweet_count, tweet.user.followers_count,tweet.user.friends_count,tweet.retweeted] for tweet in alltweets]
    
    #write the csv
    import pandas as pandas
    pd = pandas.DataFrame(outtweets)
    pd.to_csv(screen_name+"_tweets.csv")
    return alltweets[0];

onetweet=get_all_tweets("vanderbellen")
```
