```python
import pandas as pd
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
csvname='hofer_clean_done'
df = pd.read_csv(csvname+'.csv')
```

```python
df.head()
```

```python
stop = set(stopwords.words('german'))
```

```python
stop.add('aber')
stop.add('beim')
stop.add('schon')
stop.add('dass')
```

```python
# join tweets to a single string
words = ' '.join(df['text'])

# remove URLs, RTs, and twitter handles
no_urls_no_tags = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                                and word.lower() not in stop
                            ])

# Generate a word cloud image
wordcloud = WordCloud(background_color="white",width=600,height=600).generate(no_urls_no_tags)

# Display the generated image:
# the matplotlib way:
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111)
ax.imshow(wordcloud)
ax.axis("off")
```

```python
 df['hashtags'] = df['hashtags'].astype('str') 
# join tweets to a single string
words = ' '.join(df['hashtags'])

# remove URLs, RTs, and twitter handles
no_urls_no_tags = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                                and word != 'nan'
                                and word.lower() not in stop
                            ])

# Generate a word cloud image
wordcloud = WordCloud(background_color="white",width=600,height=600).generate(no_urls_no_tags)

# Display the generated image:
# the matplotlib way:
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111)
ax.imshow(wordcloud)
ax.axis("off")
```
